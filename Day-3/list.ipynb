Indexing 0 se start hoti hai.

```python
my_list = [1, "hello", 3.14, True]
list
```

üîπ 2. Python List ki Key Features

Ordered: Elements ka order important hai, same order maintain hota hai.

Mutable: Change kar sakte ho (add, remove, update).

Heterogeneous: Mixed types allowed.

Dynamic: Size automatically grow/shrink hoti hai.

Indexing & Slicing: Kisi bhi element ko index se access kar sakte ho.

# a) Access elements

```python
my_list = [10, 20, 30, 40]
print(my_list[0])   # 10
print(my_list[-1])  # 40 (last element)
```

# b) Add elements

```python
my_list.append(50)       # end mein add
my_list.insert(1, 15)    # specific index par add
my_list
```

# c) Remove elements

remove(value): Value ke basis par remove

pop(index): Index ke basis par remove (default last)

clear(): Sab elements delete kar de

```python
my_list.remove(30)  # value se remove
my_list.pop()       # last element remove
my_list.clear()     # empty list
```

# d) Modify elements

List mutable hoti hai, elements update kiye ja sakte hain:

```python
my_list[2] = 25   # 3rd element change kar diya
```

# e) Slice elements

```python
sub_list = my_list[1:3]  # index 1 to 2
print(my_list[1:4])   # [20, 30, 40] (index 1 to 3)
print(my_list[:3])    # [10, 20, 30] (start to 2)
print(my_list[2:])    # [30, 40, 50] (2 to end)
print(my_list[::2])   # [10, 30, 50] (step 2)

```

# 6Ô∏è‚É£ Search / Check Existence

in operator se check karte hain

```python
print(20 in my_list)   # True
print(50 in my_list)   # False
#index(value): Value ka index nikalta hai
print(my_list.index(25))  # 1
```

# 7Ô∏è‚É£ Length / Count

len(): Total elements

count(value): Specific value ka count

```python
print(len(my_list))       # 3
print(my_list.count(25))  # 1
```

# 8Ô∏è‚É£ Sort / Reverse

sort(): Ascending order

sort(reverse=True): Descending order

reverse(): List reverse order

```python
my_list = [40, 10, 30, 20]
my_list.sort()
print(my_list)  # [10, 20, 30, 40]
my_list.sort(reverse=True)
print(my_list)  # [40, 30, 20, 10]
my_list.reverse()
print(my_list)  # [10, 20, 30, 40]
```

# üîü Other Useful Operations

Concatenation: + operator

```python
list1 = [1,2]
list2 = [3,4]
list3 = list1 + list2  # [1,2,3,4]
# Repetition: * operator
list1 = [1,2]
list2 = list1 * 3   # [1,2,1,2,1,2]
```

| Feature           | C++/Java Array           | Python List           |

# üîπ  extend()

Ek list ke elements ko dusri list ke end mein add karta hai.

```python
list1 = [1, 2, 3]
list2 = [4, 5, 6]
list1.extend(list2)
print(list1)  # [1, 2, 3, 4, 5, 6]

# Difference from append():
# append() puri list ko ek single element ki tarah add karta hai.

list1 = [1, 2, 3]
list1.append([4,5])
print(list1)  # [1,2,3,[4,5]]
```

# üîπ min() and max()

List ke smallest aur largest value ko return karte hain.

```python
numbers = [10, 20, 5, 30]
print(min(numbers))  # 5
print(max(numbers))  # 30
```

# sum()

List ke saare numbers ka sum nikalta hai.

```python
numbers = [10, 20, 30]
print(sum(numbers))  # 60
```
