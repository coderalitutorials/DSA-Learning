# Python Tuples
## ðŸ”¹ 1. Tuple kya hai?

Tuple ek ordered collection hai, bilkul list ki tarah.

Immutable hota hai â†’ ek baar create ho gaya toh change nahi kar sakte.

Heterogeneous data allowed â†’ integer, string, float ek saath.

Indexing aur slicing possible hai.

```python
my_tuple = (1, "hello", 3.14)
print(my_tuple[0])    # 1
print(my_tuple[-1])   # 3.14
```

##  Tuple Creation

```python
t1 = (1,2,3)
t2 = ("a", "b", "c")
t3 = (1, "hello", 3.14)
t4 = ()            # empty tuple
t5 = (5,)          # single element tuple (comma zaruri hai)
# Note: Single element tuple ke liye comma lagana
# zaruri hai, warna Python usse normal value samjhega.
```

# Tuple Operations

## a) Indexing & Slicing

```python
t = (10, 20, 30, 40)
print(t[0])     # 10
print(t[-1])    # 40
print(t[1:3])   # (20,30)
```

## b) Concatenation & Repetition

```python
t1 = (1,2)
t2 = (3,4)
print(t1 + t2)   # (1,2,3,4)
print(t1 * 3)    # (1,2,1,2,1,2)
```

## c) Membership Chec

```python
t = (1,2,3)
print(2 in t)    # True
print(5 in t)    # False
```

## d) Length, min, max, sum

```python
t = (5,10,15)
print(len(t))    # 3
print(min(t))    # 5
print(max(t))    # 15
print(sum(t))    # 30
```

# e) Iteration

```python
t = (1,2,3)
for val in t:
    print(val)
```

# Tuple vs list

| Feature    | List             | Tuple      |
| ---------- | ---------------- | ---------- |
| Mutability | Mutable          | Immutable  |
| Syntax     | `[ ]`            | `( )`      |
| Methods    | Many             | Limited    |
| Use case   | Frequent changes | Fixed data |


# Python Strings
ðŸ”¹ 1. String kya hai?

String ek sequence of characters hoti hai.

Python mein string immutable hoti hai â†’ ek baar create hone ke baad change nahi ki ja sakti.

Strings ko single ' ', double " " ya triple quotes ''' ''' / """ """ me likh sakte hai

```python
s1 = "Hello"
s2 = 'Python'
s3 = """This is a multi-line string"""
```

# ðŸ”¹ 2. String Features

Ordered â†’ characters ka order maintain hota hai

Immutable â†’ change nahi ho sakta

Indexing & Slicing â†’ characters access kar sakte hain

Concatenation & Repetition possible

# ðŸ”¹ 3. Accessing Characters (Indexing & Slicing)

```python
s = "Python"
print(s[0])      # 'P'
print(s[-1])     # 'n'
print(s[0:4])    # 'Pyth'
print(s[::2])    # 'Pto' (step = 2)
```

# ðŸ”¹ 4. String Operations
a) Concatenation (+) & Repetition (*)

```python
s1 = "Hello"
s2 = "World"
print(s1 + " " + s2)  # 'Hello World'
print(s1 * 3)         # 'HelloHelloHello'
```

b) Membership (in)

```python
s = "Python"
print("Py" in s)     # True
print("Java" in s)   # False
```

c) Length (len())

```python
s = "Python"
print(len(s))        # 6
```

# ðŸ”¹ 5. Common String Methods

| Method                  | Description & Example                                                            |
| ----------------------- | -------------------------------------------------------------------------------- |
| `lower()`               | Lowercase letters <br> `"Hello".lower()` â†’ `'hello'`                             |
| `upper()`               | Uppercase letters <br> `"Hello".upper()` â†’ `'HELLO'`                             |
| `capitalize()`          | First letter capital <br> `"hello".capitalize()` â†’ `'Hello'`                     |
| `title()`               | First letter of each word capital <br> `"hello world".title()` â†’ `'Hello World'` |
| `strip()`               | Remove leading/trailing spaces <br> `"  hi  ".strip()` â†’ `'hi'`                  |
| `replace(old, new)`     | Replace substring <br> `"hi world".replace("hi","Hello")` â†’ `'Hello world'`      |
| `split(sep)`            | Convert string to list <br> `"a,b,c".split(",")` â†’ `['a','b','c']`               |
| `join(iterable)`        | Join list/tuple into string <br> `",".join(["a","b"])` â†’ `'a,b'`                 |
| `find(sub)`             | Index of substring <br> `"python".find("th")` â†’ 2                                |
| `count(sub)`            | Count occurrence <br> `"python".count("o")` â†’ 1                                  |
| `startswith(sub)`       | Check starting <br> `"python".startswith("py")` â†’ False (case-sensitive)         |
| `endswith(sub)`         | Check ending <br> `"python".endswith("on")` â†’ True                               |
| `isalpha()`             | Check letters only <br> `"Python".isalpha()` â†’ True                              |
| `isdigit()`             | Check digits only <br> `"123".isdigit()` â†’ True                                  |
| `islower() / isupper()` | Check case                                                                       |


# ðŸ”¹ 6. Iteration over String

```python
s = "Python"
for ch in s:
    print(ch)
```
