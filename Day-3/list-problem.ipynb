# Problem: Running Sum of 1D Array

```python
def running_sum(nums):
    result = []
    total = 0
    for num in nums:
        total += num
        result.append(total)
    return result

# Example
nums = [1, 2, 3, 4]
print(running_sum(nums))  # Output: [1, 3, 6, 10]
```

# Problem: Remove Duplicates from Sorted List

```python
def remove_duplicates(nums):
    if not nums:
        return []
    
    result = [nums[0]]  # first element always included
    for num in nums[1:]:
        if num != result[-1]:  # compare with last element in result
            result.append(num)
    return result

# Example
nums = [1, 1, 2, 3, 3, 4]
print(remove_duplicates(nums))  # Output: [1, 2, 3, 4]
```
